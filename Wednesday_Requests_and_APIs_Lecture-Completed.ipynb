```python
class CityWeather:
    def __init__(self, name, region, current, feels_like, condition):
        self.city = name
        self.region = region
        self.current_temp = current
        self.feels_like = feels_like
        self.condition = condition
        
    def __str__(self):
        return f"It is currently {self.condition} and {self.current_temp} in {self.city}, {self.region}. \
It feels like {self.feels_like}"
    
    def __repr__(self):
        return f"<CityWeather|{self.city}>"
```

```python
class WeatherAPI:
    base_url = "http://api.weatherapi.com/v1"
    
    def __init__(self, api_key):
        self.api_key = api_key
        
    def _get(self, city, api_method, **params):
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={city}"
        if params:
            for key, value in params.items():
                request_url += f"&{key}={value}"
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print('There was an error!')
            
    def get_current_weather(self, city):
        weather_data = self._get(city, '/current.json')
        if weather_data:
            current_temp = weather_data['current']['temp_f']
            feels_like = weather_data['current']['feelslike_f']
            current_condition = weather_data['current']['condition']['text']
            city_name = weather_data['location']['name']
            region_name = weather_data['location']['region']
            
            city_obj = CityWeather(city_name, region_name, current_temp, feels_like, current_condition)
            return city_obj
        
    def get_forecast_weather(self, city, days=1):
        weather_data = self._get(city, '/forecast.json', days=days)
        return weather_data   
```

```python
client = WeatherAPI('1ac9901266fa4ab48f9185441220510')
```

```python
chicago = client.get_current_weather('Chicago')
print(chicago)
print(chicago.feels_like)
```

```python
london = client.get_current_weather('London')
print(london)
```

```python
new_york = client.get_current_weather('New York')
print(new_york)
```
